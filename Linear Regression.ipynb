```python
# Import Libary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# New Section

```python
df=pd.read_csv("/content/headbrain.csv")
df
```

```python
df.describe()
```

```python
# Declare dependent variable(Y) and independent variable(X)

X=df['Head Size(cm^3)'].values
Y = df['Brain Weight(grams)'].values
```

```python
print(X, Y)
```

```python
# Plot the Input Data
plt.scatter(X, Y, c='green', label='Data')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Calculating coefficient

# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate theta1 and theta2
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
m = numer / denom
print(m)
c = mean_y - (m * mean_x)

# Printing coefficients
print("coefficients for regression",m, c)
```

```python


plt.rcParams['figure.figsize'] = (20.0, 10.0)
# max_x = np.max(X) + 100
# min_x = np.min(X) - 100

y = c + m * X

# Ploting Line
plt.plot(X, y, color='blue', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='green', label='Scatter data')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Calculating Root Mean Squares Error
rmse = 0
for i in range(n):
    y_pred = c + m * X[i]
    rmse += (Y[i] - y_pred) ** 2

rmse = np.sqrt(rmse/n)
print("Root Mean Square Error is",rmse)
```

```python
# Calculating R2 Score
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = c + m * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score",r2)
```
